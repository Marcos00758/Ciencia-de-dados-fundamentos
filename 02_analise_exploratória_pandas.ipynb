```python
# Célula de importação

import pandas as pd
```

```python
# Célula de dados(nome, idade, nota e faltas de alunos)

dados = {
    "nome": ["Ana", "Bruno", "Carlos", "Diana", "Edu"],
    "idade": [20, 17, 22, 19, 16],
    "nota": [8.5, 5.0, 6.8, 9.2, 4.5],
    "faltas": [2, 5, 1, 0, 6]
}

df = pd.DataFrame(dados)
df
```

```python
# Extração de informações

df.info()
df.describe()
```

```python
# Quantos alunos são maiores de idade?
maiores = []
for i in df['idade']:
    if i >= 18:
        maiores.append(i)
print(len(maiores))
# ou
# df[df['idade']>=18].shape[0]
```

```python
# Qual a média da nota dos aprovados
aprovados = df[df['nota']>=6]
aprovados['nota'].mean()
```

```python
# Quem tem mais faltas:
df.loc[df['faltas'].idxmax()]["na"]
```

```python
# Criação de nova coluna
def classificar_aluno(linha):
    if linha['nota']>=6 and linha['faltas']<=3:
        return "Aprovado"
    else:
        return "Reprovado"
df['Status']=df.apply(classificar_aluno, axis=1)
df
```
