# Conceitos

**Fila(FIFO):** Primeiro a entrar é o primeiro a sair
**Pilha(LIFO)** Último a entrar é o primeiro a sair

Essas estruturas são fundamentais para organização e execução de tarefas em sistemas computacionais


```python
def ordenar_tarefas(tarefas):
    fila = []
    pilha = []

    for tarefa in tarefas:
        if tarefa["urgente"]:
            pilha.append(tarefa["nome"])
        else:
            fila.append(tarefa["nome"])

    return fila, pilha


def executar_tarefas(fila, pilha):
    print("Execução:")

    while pilha:
        print(pilha.pop())

    while fila:
        print(fila.pop(0))
```

```python
tarefas = [
    {"nome": "backup", "urgente": False},
    {"nome": "atualizar_sistema", "urgente": True},
    {"nome": "limpar_cache", "urgente": False},
    {"nome": "reiniciar_servidor", "urgente": True},
    {"nome": "verificar_logs", "urgente": False}
]

fila, pilha = ordenar_tarefas(tarefas)
executar_tarefas(fila, pilha)
```

```python
## Conclusão

O uso de fila e pilha permite controlar a ordem de execução das tarefas.
Tarefas urgentes foram tratadas com prioridade utilizando pilha, enquanto
tarefas normais foram organizadas em fila.
```
